```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:.6f}'.format
df = pd.read_csv('약국별속성최종본.csv',encoding='EUC-KR')
```

```python
df_2 = df[df.columns[6:]]
col_lst=list(df_2.columns)
```

```python
scaler = MinMaxScaler()
df_scaled_minmax=pd.DataFrame()
df_scaled_minmax[col_lst] =scaler.fit_transform(df_2)
```

```python
df_scaled_minmax
```

```python
df_scaled_minmax['잠재구매']= df_scaled_minmax['경증환자수']+\
df_scaled_minmax['경증환자_약품비']+ \
df_scaled_minmax['경증_약품비_비율']+ \
df_scaled_minmax['경증_환자수_비율']+ \
df_scaled_minmax['반경내_생활인구_추정']+\
df_scaled_minmax['평균유동인구']+\
df_scaled_minmax['반경내_인구수_합']

#가중치를 곱해줌
df_scaled_minmax['잠재구매지수']=df_scaled_minmax['잠재구매']*0.3
```

```python
df_scaled_minmax['교통접근']= df_scaled_minmax['버스개수'] +\
df_scaled_minmax['지하철개수']+\
df_scaled_minmax['bus_ratio']+\
df_scaled_minmax['sub_ratio']

df_scaled_minmax['교통접근지수']=df_scaled_minmax['교통접근']*0.5
```

```python
df_scaled_minmax['잠재경쟁']=df_scaled_minmax['store_ratio']+\
df_scaled_minmax['nd_ratio']+df_scaled_minmax['emer_ratio']+\
df_scaled_minmax['응급실개수']+\
df_scaled_minmax['심야약국개수']+\
df_scaled_minmax['편의점개수']

df_scaled_minmax['잠재경쟁지수']=df_scaled_minmax['잠재경쟁']*0.2
```

```python
df_score = df_scaled_minmax.copy()

df_score = df_score[['잠재구매지수','교통접근지수','잠재경쟁지수']]
```

```python
df_score['총점']=(df_score['잠재구매지수'])+ \
(df_score['교통접근지수'])+ \
(df_score['잠재경쟁지수'])
df_score['순위'] = df_score['총점'].rank(ascending=False)
df_score
```

```python
df_fin = pd.concat([df[df.columns[0:6]],df_score],axis=1)
df_fin.sort_values('순위')
```

```python
df_fin.to_csv('최종산출데이터.csv',encoding='euc-kr',index=False)
```
